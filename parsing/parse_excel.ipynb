```python
import pandas as pd
import numpy as np

# Load Excel to DataFrame
filename = 'Расписания/ОН_Магистратура_ 1 курс ФТКМ (1).xls'
df = pd.read_excel(
    filename,
    usecols="E:Z",
    skiprows=5,
    skipfooter=5,
    header=None,
    keep_default_na=False
)

#with pd.option_context('display.max_rows', None,
#                       'display.max_columns', None,
#                       'display.precision', 3,
#                       ):
#    display(df)
df[5][0]='пара'
df[4][0]='день недели'
df
```

```python
df2 = pd.DataFrame()
df2[4]=df[4]
df2[5]=df[5]
for i in range (6,df.shape[1]+5-1,4):
    df2[i]=df[[i,i+1,i+2,i+3]].astype(str). agg(''. join , axis= 1 )
```

```python
#with pd.option_context('display.max_rows', None,
#                       'display.max_columns', None,
#                       'display.precision', 3,
#                       ):
#    display(df2)
df2
```

```python
df3 = pd.DataFrame(data=df2[1:])
df3.columns=df2[:1].values[0]
df3
```

```python
result = df3.groupby(np.arange(len(df3))//3).sum()
result
```

```python
with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    display(result)
```

```python
if (".xlsx" in filename):
    result.to_json("json/"+filename[:-5]+".json",orient="split")
else:
    result.to_json("json/"+filename[:-4]+".json",orient="split")
```
